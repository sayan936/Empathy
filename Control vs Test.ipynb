```python
import os
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
test = pd.read_csv('Test_group.csv')
```

```python
test.drop('Unnamed: 0',axis = 1)
```

```python
control = pd.read_csv('Control_group.csv')
```

```python
control.drop('Unnamed: 0',axis = 1)
```

### Plotting mean pupil diameter test vs control

```python


df = pd.DataFrame({'Test': test['Mean Pupil Diameter'], 'Control': control['Mean Pupil Diameter']})

# Plot a box plot
sns.boxplot(data=df)
plt.ylabel('Mean Pupil Diameter')
```

### Plotting Average Empathy Score Test vs Control

```python
df = pd.DataFrame({'Test': test['Avg_Empathy_Score'], 'Control': control['Avg_Empathy_Score']})

# Plot a box plot
sns.boxplot(data=df)
plt.ylabel('Avg_Empathy_Score')
```

### Difference in empathy score 

```python
diff_empathy_score = test['Avg_Empathy_Score'] - control['Avg_Empathy_Score']

# Plot a line chart
plt.plot(diff_empathy_score)
plt.xlabel('Index')
plt.ylabel('Difference in Empathy Score')
```
